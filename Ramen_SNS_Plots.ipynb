```python
import pandas as pd 
import seaborn as sns 
```

```python
df_ramen=pd.read_csv('ramen-ratings.csv')
```

1.Read the data from the CSV file into a DataFrame and display the first five rows 

```python
df_ramen.head()
```

2.Rename the Packacking column to Packaging and the Rating column to Rating

```python
df_ramen.rename(columns={'Style':'Packaging','Stars':'Rating'},inplace=True)
df_ramen.head()
```

3.Create a bar plot to show the Rating by Packaging with a Seaborn specific method.

```python
sns.barplot(data=df_ramen,x='Packaging',y='Rating',errorbar=None)
```

4.Create a bar plot to show the Rating by Packacking with a Seaborn generic method.

```python
sns.catplot(data=df_ramen,kind='bar',
            x='Packaging',y='Rating',errorbar=None)
```

5.Modify the plot you just created to use a custom title and x-axis label.

```python
g=sns.catplot(data=df_ramen,kind='bar',
            x='Packaging',y='Rating',errorbar=None)
for ax in g.axes.flat:
    ax.set(title='This Show The Rating By Packaging',xlabel='Type of packaging')
    
```

6.Use the query()method to filter the data to contain only data for Japan, India,
 Taiwan, and the U.S. Plot the stars by style for each country on 
 a subplot with two subplots per row.

```python
df_ramen.Country.replace({'USA':'U.S.'},inplace=True)
```

```python
g=sns.catplot(data=df_ramen.query("Country in ['Japan', 'India', 'Taiwan', 'U.S.']")
            ,kind='bar',hue='Packaging',
            x='Packaging',y='Rating',col='Country',col_wrap=2,legend=False,
            height=3,aspect=1.2)
```

7.Modify the plot you just created to have a custom title for each subplot
. Hint: Use the enumerate() method

```python
g=sns.catplot(data=df_ramen.query("Country in ['Japan', 'India', 'Taiwan', 'U.S.']"),
            kind='bar',hue='Packaging', x='Packaging',y='Rating',col='Country',
            col_wrap=2,legend=False,height=3,aspect=1.2)
country=df_ramen['Country'].drop_duplicates().to_list()
for index ,ax in enumerate(g.axes.flat):
    ax.set_title(country[index])
```
